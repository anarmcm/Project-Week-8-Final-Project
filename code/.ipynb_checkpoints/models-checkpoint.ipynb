# Logistic Regression: Bonus: Good or bad movie?

Classifying movies according to their vote_average to see if they're good or not. If the average is higher than 6.5 then the movie is good, if not, the movie is bad. 

```python
vote_list = list(data_vote.vote_average)
```

```python
vote_dummy = []
for i in vote_list: 
    if i > 6.5: 
        vote_dummy.append(1)
    else:
        vote_dummy.append(0) 
```
